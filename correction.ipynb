```python
import pandas as pd
```

```python
ds = pd.read_csv("data/master_dataset.csv")
```

```python
ds.shape
```

```python
df = pd.read_csv("data/dataset_lyrics_popularity.csv")
```

```python
df2_subset = df[['artist', 'title', 'lyrics', 'happiness_from_lyrics', 'popularity']].drop_duplicates(subset=['artist', 'title'])

# Merge avec df1
df_merged = pd.merge(
    ds,
    df2_subset,
    on=['artist', 'title'],
    how='left'
)

```

```python
df_merged.shape
```

```python
df_merged.isna().sum()
```

```python
df_merged.to_csv("data/dataset_lyrics_popularity.csv", index=False)
```
